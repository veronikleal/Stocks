```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
stock = ['AMZN']
stock_data = yf.download(stock, start = '2024-12-10', end ='2024-12-11', interval =  '5m')
print(stock_data.head())
data = stock_data.loc[:, 'Close'].copy()
```

```python
data.plot(figsize = (17,8), fontsize = 18)
plt.show()
```
